```python
newdf.loc[0,0]=58
# it add the new column in the dataframe
```

```python
newdf
```

```python
# if we want to print some specific location data
newdf.loc[[1,2],['C','D']]
# it return a copy of the dataframe
```

```python
# if we want all the row and some specific column
newdf.loc[:,['C','D']]
```

```python
# if we want all the column and some specific rows
newdf.loc[[1,2],:]
```

```python
# if we want all the row which have value less than 0.2
newdf.loc[(newdf['B']<0.2)]
```
